## Header information

Metadata from the LAS file's header is stored in an ``OrderedDictionary`` attribute, one for each section (``version``, ``well``, ``curves``, and ``params``). ``OrderedDictionary`` objects are just what they sound like: very similar to a normal Python dictionary.

Each entry in the header sections is stored by *mutable named tuple* called a ``HeaderItem``, which is created using the ``namedlist`` library. It's very straightforward:

```python
from lasio.las import HeaderItem

example = HeaderItem(mnemonic="ROP", unit="m/min", value=2.3, descr="Rate of penetration")

print(example)

print(example.mnemonic)
print(example.unit)
print(example.value)
print(example.descr)
```

In order from top to bottom as they usually are in LAS files, the header sections are:

```python
l.version
```

```python
l.version['VERS'].value
```

```python
l.well
```

To change metadata information you have to edit the relevant field of the ``HeaderItem`` object:

```python
l.well['SRVC'].value = "My logging company -- not anyone else's"
print(l.well["SRVC"])
```

The other section with header items is very similar...

```python
l.params
```

For convenience there's a special attribute which brings together all the header metadata mnemonics and values

```python
l.metadata
```

The ``~Other`` section is left as raw text.

```python
print(l.other)
```

Note that commented-out lines beginning with ``#`` elsewhere in the file are not read in at all at the moment (this will be fixed in the future).

## Curves (more detail)

The curve definitions in the ``~Curve`` section are stored as ``Curve`` objects. The only difference between a ``Curve`` object and a ``HeaderItem`` is that the former have an additional ``data`` field:

```python
l.curves
```

```python
neutron_log = l.get_curve("NEUT")
print(neutron_log)
```

Notice of course that whereas ``l["NEUT"]`` returns you a ``numpy.ndarray`` directly, using the ``get_curve`` method gives you the whole ``Curve`` object, with the unit, value (usually an API curve type code), and descr attributes as well as the data.

The reference/index curve of the LAS file (e.g. the depth or time column) can be accessed directly by the ``.index`` attribute:

```python
print(l["DEPT"])
print(l.index)
```

## pandas.DataFrame attribute

If [pandas](http://pandas.pydata.org/) is installed, lasio will automatically create a pandas DataFrame attribute for the las object.  

It's a fully functional pandas DataFrame object:

```python
type(l.df)
```

View a snapshot of the dataframe with the `head()` method, which takes number of lines as an argument:

```python
l.df.head(10)
```

Use the `describe()` method to return basic statistical information about the curves in the dataframe:

```python
l.df.describe()
```

Calculations are easy in a dataframe.  As a simple example, let's convert conductivity to resistivity:

```python
l.df['RES'] = 1000 / l.df.COND
```

```python
l.df[100:100.5]
```
