# reading some real-life data from a LAS 2.0 file

```python
%matplotlib inline
import las_reader
```

```python
log = las_reader.LASFile(r"F:\code\las\examples\2.0\D-D' LAS Files\Bean\Bean_C.las")
```

There are 11,201 rows of data:

```python
log.data
```

And a primitive plot of the two series:

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 3))
ax = fig.add_subplot(111)
ax2 = ax.twinx()

log.data.GR.plot(ax=ax, color='b', grid=False, label='GR')
log.data.RHOB.plot(ax=ax2, color='r', grid=False, label='RHOB')

ax.legend(loc='upper right')
ax2.legend(loc='lower right')
```
